```python
fig, ax = plt.subplots()
for imax in range(10):
    arr = [1,1,1]+[3<<i for i in range(imax)]
    ha = H(arr)
    fldr_toll, ky_toll = get_tree_entropy(gen_fldr_tree(arr)) - ha, get_tree_entropy(gen_ky_tree(arr)) - ha
    print(f"Depth {imax+2:2}: FLDR toll: {fldr_toll:.10f} bits, KY toll: {ky_toll:.10f} bits, difference: {fldr_toll - ky_toll:.10f} bits")
    plot_tree_tolls(arr, gen_tree=gen_fldr_tree, ax=ax)
```

```python
fig, ax = plt.subplots()
# set logarithmic scale for x-axis
ax.set_xscale('log')
# make plot huge
fig.set_size_inches(20, 10)
# add title and axis labels
ax.set_title(r'FLDR Interpolated Tree Tolls for $[1,2^n,(1+2^n),(1+2^n)*2,\dots]$')
ax.set_xlabel('Depth')
ax.set_ylabel('Tree Toll (bits)')
for imax in range(1,20):
    pow2nplus1 = (1<<imax) + 1
    arr = [1, pow2nplus1-1] + [pow2nplus1<<i for i in range(1,imax+1)]
    ha = H(arr)
    fldr_toll, ky_toll = get_tree_entropy(gen_fldr_tree(arr)) - ha, get_tree_entropy(gen_ky_tree(arr)) - ha
    print(f"Depth {imax+2:2}: FLDR toll: {fldr_toll:.10f} bits, KY toll: {ky_toll:.10f} bits, difference: {fldr_toll - ky_toll:.10f} bits")
    plot_tree_tolls(arr, gen_tree=gen_fldr_tree, ax=ax)
```

```python
fig, ax = plt.subplots()
# set logarithmic scale for x-axis
ax.set_xscale('log')
# make plot huge
fig.set_size_inches(6,4)
# add title and axis labels
ax.set_title(r'FLDR Interpolated Tree Tolls for [1, $p-1$, $p$, $2p$, $4p$, $\dots$, $2^{\lfloor\log p\rfloor}p$] ($p = 1669$)')
ax.set_xlabel('Depth')
ax.set_ylabel('Tree Toll (bits)')
for p in primes[100:101]:
    pow2nplus1 = (1<<imax) + 1
    arr = [1, p-1] + [p<<i for i in range(p.bit_length())]
    ha = H(arr)
    fldr_toll, ky_toll = get_tree_entropy(gen_fldr_tree(arr)) - ha, get_tree_entropy(gen_ky_tree(arr)) - ha
    print(f"Depth {imax+2:2}: FLDR toll: {fldr_toll:.10f} bits, KY toll: {ky_toll:.10f} bits, difference: {fldr_toll - ky_toll:.10f} bits")
    plot_tree_tolls(arr, gen_tree=gen_fldr_tree, toll_cutoff = lambda *args:False, ax=ax)
plt.show()
# Depth 21: FLDR toll: 2.4529622941 bits, KY toll: 0.0009730088 bits, difference: 2.4519892853 bits
```
