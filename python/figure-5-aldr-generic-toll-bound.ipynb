```python
dk_max = 12
dkr = range(dk_max+1)
def rejection_toll_bound(dk):
    return ((2**dk + 1) / 2**dk) * (2 + H([1, 2**dk]))

fig, ax = plt.subplots(dpi=150)

ax.set_xlabel(r'$\delta = K-k$')
ax.set_ylabel(r'Entropy Cost (bits)')
ax.set_title("Upper Bounds on Entropy Cost of ALDR Sampler")
tolls = [(rejection_toll_bound(dk)) for dk in dkr]
ax.plot(dkr, tolls, label=r'$\displaystyle\frac{2^{\delta}+1}{2^{\delta}} \Bigg[2+H_{\rm b}(1/(2^{\delta}+1))\Bigg]$', linestyle='-', color='k')
ax.plot(dkr, [2 + (4 + dk)/2**dk  for dk in dkr], label=r'$\displaystyle 2+\frac{4+\delta}{2^{\delta}}$', linestyle='dotted', color='k')
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.1))
ax.grid(which='both', alpha=0.25)
ax.legend(framealpha=1, fancybox=False, fontsize=10)

fig.set_tight_layout(True)
fig.set_size_inches(6, 4)
fig.savefig(output_dir / 'aldr-generic-toll-bound.png')
```
