### Plots of relative tolls and their decomposition into $\nu$ and $H$.

```python
from customtree import *
```

```python
denom_pow_max = 14
plt.gcf().set_size_inches(6, 4)
# plt.xscale('log', base=2)
plt.gca().xaxis.set_major_locator(ticker.LogLocator(base=2, subs=[x/4 for x in range(4,9)]))
plt.gca().xaxis.set_minor_locator(ticker.LogLocator(base=2, subs=[x/8 for x in range(9,16)]))
plt.title('Relative Tolls')
plt.xlabel(r'$x$')
plt.ylabel(r'$(\nu(x)-H(x))/x$')
for denom_pow in range(1, denom_pow_max + 1):
    denom = 1 << denom_pow
    # use rainbow based on denom_pow
    color = plt.cm.rainbow(1 - denom_pow / denom_pow_max)
    for v in range(denom//2+1, denom, 2):
        nu = 0
        for i in range(1, denom_pow + 1):
            if v & (1<<(denom_pow - i)):
                nu += i / (1<<i)
        nu_max = nu + 2 / denom
        x = v / denom
        H = - x * math.log2(x)
        plt.plot([v / (1 << denom_pow)]*2, [(nu-H)/x, (nu_max-H)/x], color=color, linewidth=10/denom_pow)
plt.show()
```

```python
denom_pow_max = 14
plt.gcf().set_size_inches(15, 10)
# plt.xscale('log', base=2)
plt.gca().xaxis.set_major_locator(ticker.LogLocator(base=2, subs=[x/4 for x in range(4,9)]))
plt.gca().xaxis.set_minor_locator(ticker.LogLocator(base=2, subs=[x/8 for x in range(9,16)]))
plt.title('Relative Tolls')
plt.xlabel(r'$x$')
plt.ylabel(r'$(\nu(x)-H(x))/x$')
for denom_pow in range(1, denom_pow_max + 1):
    denom = 1 << denom_pow
    # use rainbow based on denom_pow
    color = plt.cm.rainbow(1 - denom_pow / denom_pow_max)
    for v in range(denom//2+1, denom, 2):
        nu = 0
        for i in range(1, denom_pow + 1):
            if v & (1<<(denom_pow - i)):
                nu += i / (1<<i)
        nu_max = nu + 2 / denom
        x = v / denom
        H = - x * math.log2(x)
        plt.plot([v / (1 << denom_pow)]*2, [(nu-H)/x, (nu_max-H)/x], color=color, linewidth=10/denom_pow)
        plt.plot([v / (1 << denom_pow)]*2, [nu/x, nu_max/x], color='black', linewidth=1, marker='.')
        plt.plot([v / (1 << denom_pow)], [-H/x], color='red', linewidth=1, marker='.')

plt.plot([], [], color='black', linewidth=1, marker='.', label=r'$\nu(x)/x$')
plt.plot([], [], color='red', linewidth=1, marker='.', label=r'$-H(x)/x$')
plt.legend()
plt.show()
```
