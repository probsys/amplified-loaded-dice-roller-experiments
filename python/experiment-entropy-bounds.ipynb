This is a plot of the binary entropy bounds proved by Tops√∏e. The log bounds are tight for small $p$, while the power bounds are tight for $p \approx 1/2$. The theorems bounding the toll of ALDR require the even tighter bound of $(1/\ln(2)-\log_2(p))p$.

```python
import math
import matplotlib.pyplot as plt
import sympy
```

```python
p = [i/100 for i in range(1,100)]
H = [-p * math.log2(p) - (1-p) * math.log2(1-p) for p in p]
# log bounds
H_lower_log = [math.log(2) * math.log2(p) * math.log2(1-p) for p in p]
H_upper_log = [math.log2(p) * math.log2(1-p) for p in p]
# power bounds
H_lower_pow = [4 * p * (1-p) for p in p]
H_upper_pow = [(4 * p * (1-p))**(1/math.log(4)) for p in p]
# my bound
H_my = [(1/math.log(2)-math.log2(p))*p for p in p]
fig, ax = plt.subplots()
ax.set_xlabel('$p$')
ax.set_ylabel('Binary Entropy')
ax.set_title("Tops√∏e's Binary Entropy Bounds")
ax.plot(p, H, label=r'$H_{\rm b}(p)$')
ax.plot(p, H_lower_log, label=r'$\ln(2) \log(p) \log(1-p)$')
ax.plot(p, H_upper_log, label=r'$\log(p) \log(1-p)$')
ax.plot(p, H_lower_pow, label='$4p(1-p)$')
ax.plot(p, H_upper_pow, label=r'$(4p(1-p))^{1/\ln(4)}$')
ax.plot(p, H_my, label=r'$(1/\ln(2)-\log_2(p))p$')
ax.legend()
plt.show()
```
